# 4. Distribusi Harga Produk

```python
plt.figure(figsize=(8,5))

plt.hist(df["price_avg"], bins=30)

plt.title("Distribution of Product Price")

plt.xlabel("Price")

plt.ylabel("Frequency")

plt.show()
```

## Insight

Grafik menunjukkan distribusi harga produk pada dataset.

# 5. Distribusi Rating Produk

```python
plt.figure(figsize=(8,5))

plt.hist(df["rating"], bins=20)

plt.title("Distribution of Rating")

plt.xlabel("Rating")

plt.ylabel("Frequency")

plt.show()
```

# 6. Distribusi Jumlah Review

```python
plt.figure(figsize=(8,5))

plt.hist(df["review_count"], bins=30)

plt.title("Distribution of Review Count")

plt.xlabel("Review")

plt.ylabel("Frequency")

plt.show()
```

# 7. Distribusi Jumlah Penjualan

```python
plt.figure(figsize=(8,5))

plt.hist(df["sold"], bins=30)

plt.title("Distribution of Sold Product")

plt.xlabel("Sold")

plt.ylabel("Frequency")

plt.show()
```

# 8. Distribusi Persentase Diskon

```python
plt.figure(figsize=(8,5))

plt.hist(df["discount_percent"], bins=20)

plt.title("Distribution of Discount")

plt.xlabel("Discount (%)")

plt.ylabel("Frequency")

plt.show()
```

# 9. Distribusi Keyword

```python
keyword = df["keyword"].value_counts()

keyword.plot(kind="bar")

plt.title("Number of Products by Keyword")

plt.ylabel("Total Product")

plt.show()
```

# 10. Distribusi Kategori Produk

```python
category = df["category_2"].value_counts()

category.plot(kind="bar")

plt.title("Product Category")

plt.ylabel("Total Product")

plt.show()
```

# 11. Top 10 Seller Berdasarkan Penjualan

```python
seller = (
    df.groupby("seller_name")["sold"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

seller.plot(kind="bar")

plt.title("Top 10 Seller")

plt.ylabel("Total Sold")

plt.show()
```

# 12. Outlier Harga Produk

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["price_avg"])

plt.title("Boxplot Price")

plt.show()
```

# 13. Outlier Jumlah Penjualan

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["sold"])

plt.title("Boxplot Sold")

plt.show()
```

# 14. Scatter Plot Harga vs Penjualan

Scatter plot digunakan hanya untuk melihat pola hubungan antar variabel secara visual.

Analisis korelasi akan dilakukan pada notebook berikutnya.

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["price_avg"],
    df["sold"],
    alpha=0.5
)

plt.xlabel("Price")

plt.ylabel("Sold")

plt.title("Price vs Sold")

plt.show()
```
